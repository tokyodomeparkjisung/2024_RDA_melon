네트형성초기 이미지 로딩

```python
from pathlib import Path
from PIL import Image

directory = Path(r"C:\Users\Jason\Downloads\drive-download-20240310T140536Z-001\네트형성초기")

def load_image_dataset(path):
    image_dataset = []
    allowed_extensions = ['.jpg', '.jpeg', '.png']  # 허용된 이미지 확장자들

    for file in path.iterdir():
        if file.is_file() and file.suffix.lower() in allowed_extensions:
            try:
                img = Image.open(file)
                image_dataset.append(img)
            except Exception as e:
                print(f"Error loading image: {str(e)}")

    return image_dataset

image_dataset = load_image_dataset(directory)
print(f"Loaded {len(image_dataset)} images.")
```

```python
import matplotlib.pyplot as plt

def show_image_grid(image_dataset):
    num_images = len(image_dataset)
    grid_size = int(num_images ** 0.5) + 1  # 이미지 그리드의 크기 계산

    fig, axes = plt.subplots(grid_size, grid_size, figsize=(10, 10))
    fig.subplots_adjust(hspace=0.3, wspace=0.3)

    for i, ax in enumerate(axes.flat):
        if i < num_images:
            ax.imshow(image_dataset[i])
        ax.axis('off')
    plt.suptitle('Image Dataset', size=16)
    plt.show()

# 이미지 데이터셋 전체를 그리드 형식으로 표시
if len(image_dataset) > 0:
    show_image_grid(image_dataset)
else:
    print("No images found in the dataset.")
```

```python
def show_image(image):
    plt.imshow(image)
    plt.axis('off')
    plt.title('Sample Imageset')
    plt.show()

# 이미지 데이터셋에서 랜덤하게 하나의 이미지 선택
if len(image_dataset) > 0:
    random_image = image_dataset[0]
    show_image(random_image)
else:
    print("No images found in the dataset.")
```
